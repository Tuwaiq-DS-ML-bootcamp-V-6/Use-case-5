```python
duplicated_rows = jadarat_df[jadarat_df.duplicated(keep=False)]
print(duplicated_rows)
```

```python
jadarat_df.drop_duplicates(inplace=True)
```

```python
jadarat_df.isnull().sum().sort_values(ascending = False)
```

```python
jadarat_df['qualif'].fillna(jadarat_df['qualif'].mode()[0], inplace=True)
jadarat_df['eco_activity'].fillna(jadarat_df['eco_activity'].mode()[0], inplace=True)
jadarat_df['comp_size'].fillna(jadarat_df['comp_size'].mode()[0], inplace=True)
```

```python
jadarat_df.rename(columns={'benefits': 'salary'}, inplace=True)

jadarat_df['salary'] = jadarat_df['salary'].str.replace('Salary', '').str.replace("'", '').str.replace('[.', '').str.replace(']', '')

print(jadarat_df['salary'].head())
```

```python
jadarat_df['positions'] = jadarat_df['positions'].replace({'0': ''})

print(jadarat_df['positions'].head())
```
